### Def. of Group

**Def. of Group**

A collection of elements A, B, C,... from a group when the following four conditions are satisfied:

- Closure: If $A, B \in G$, then $AB \in G$
- Associative law: $(AB)C=A(BC)$ for all $A, B, C \in G$
- Identity Element (unit element) $E$: $AE=EA=A$ for all $A \in G$
- Inverse Element $A^{-1}$: $AA^{-1}=A^{-1}A=E$

<font color='blue'>Note:</br>
$AB\neq BA$ (not commute)</br>
$AB=BA$ for all $A, B \in G$ (Abelian group)<font>
